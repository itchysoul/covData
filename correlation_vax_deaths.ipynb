```python
import pandas as pd


vax = pd.read_csv(r'covid19_vaccinations_in_the_united_states.csv', skiprows=2 )

stats = pd.read_csv(r'cov_data.csv', thousands=',', skiprows=1)

df = vax.merge(stats, on='State')

df.columns

df = df.loc[:,['Percent of Total Pop Fully Vaccinated by State of Residence',
      'Percent of 18+ Pop Fully Vaccinated by State of Residence',
         'Percent of 65+ Pop Fully Vaccinated by State of Residence',
         'deaths/ 1M pop']].apply(pd.to_numeric)

df.corr()
```
